```python
import pandas as pd
from pathlib import Path
from torch.utils.data import Dataset, DataLoader
import torch
from torchvision import transforms
from torchvision.models.segmentation import deeplabv3_resnet50
from PIL import Image
import torch.nn as nn
import matplotlib.pyplot as plt

# ==============================
# 1. Dataset 정의
# ==============================
class SkinDataset(Dataset):
    def __init__(self, dataframe, img_dir, transform=None):
        self.df = dataframe
        self.img_dir = Path(img_dir)
        self.transform = transform
        self.label_map = {"Benign":0, "Malignant":1}

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_path = self.img_dir / f"{row['isic_id']}.jpg"
        img = Image.open(img_path).convert("RGB")
        label = self.label_map[row["diagnosis_1"]]
        if self.transform:
            img = self.transform(img)
        return img, torch.tensor(label), str(img_path)

# ==============================
# 2. 모델 정의
# ==============================
class MultiTaskDeepLab(nn.Module):
    def __init__(self, num_seg_classes: int = 3, num_cls_classes: int = 2, freeze_until: str = "layer3"):
        super().__init__()
        self.seg_model = deeplabv3_resnet50(weights="COCO_WITH_VOC_LABELS_V1", aux_loss=True)
        in_ch = self.seg_model.classifier[-1].in_channels  
        self.seg_model.classifier[-1] = nn.Conv2d(in_ch, num_seg_classes, kernel_size=1)
        self.backbone = self.seg_model.backbone  

        if freeze_until:
            for name, p in self.backbone.named_parameters():
                p.requires_grad = True
                # freeze logic
                if name.split(".")[0] in ["conv1", "bn1", "relu", "maxpool", "layer1", "layer2"]:
                    if freeze_until in ("layer1", "layer2") and name.startswith(freeze_until):
                        p.requires_grad = True
                    else:
                        if freeze_until in ("layer3", "layer4"):
                            if name.startswith("layer3") or name.startswith("layer4"):
                                p.requires_grad = True
                            else:
                                p.requires_grad = False
                        elif freeze_until == "layer2":
                            if name.startswith("layer3") or name.startswith("layer4"):
                                p.requires_grad = True
                            else:
                                p.requires_grad = False
                        elif freeze_until == "layer1":
                            if name.startswith("layer2") or name.startswith("layer3") or name.startswith("layer4"):
                                p.requires_grad = True
                            else:
                                p.requires_grad = False
                        elif freeze_until == "conv1":
                            p.requires_grad = True

        self.cls_gap = nn.AdaptiveAvgPool2d((1, 1))
        self.cls_head = nn.Sequential(
            nn.Flatten(),
            nn.Linear(2048, 512),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(512, num_cls_classes),
        )

    def forward(self, x):
        seg_out = self.seg_model(x)["out"]
        feats = self.backbone(x)["out"]
        pooled = self.cls_gap(feats)
        cls_logits = self.cls_head(pooled)
        return seg_out, cls_logits

def load_checkpoint(model, path, map_location=None):
    ckpt = torch.load(path, map_location=map_location)
    model.load_state_dict(ckpt["model"], strict=False)

# ==============================
# 3. CSV 불러오기 및 전처리
# ==============================
df = pd.read_csv('./challenge-2016-test_metadata_2025-09-01.csv')
valid_labels = ["Benign", "Malignant"]
df = df.dropna(subset=["diagnosis_1"])
df = df[df["diagnosis_1"].isin(valid_labels)].reset_index(drop=True)

transform = transforms.Compose([
    transforms.Resize((512, 512)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

dataset = SkinDataset(df, img_dir="../ISIC_dataset/img", transform=transform)
loader = DataLoader(dataset, batch_size=8, shuffle=False)

# ==============================
# 4. 모델 불러오기
# ==============================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = MultiTaskDeepLab(num_seg_classes=2, num_cls_classes=2, freeze_until=None).to(device)
load_checkpoint(model, "../runs_multitask_1000_ver2/best_multitask.pt", map_location=device)
model.eval()

# ==============================
# 5. 예측 + Loss 계산 + 틀린 이미지 확인
# ==============================
cls_criterion = nn.CrossEntropyLoss()

all_preds = []
all_labels = []
wrong_images = []
total_cls_loss = 0

with torch.no_grad():
    for imgs, labels, paths in loader:
        imgs = imgs.to(device)
        labels = labels.to(device)
        seg_out, cls_logits = model(imgs)

        # Classification Loss 계산
        cls_loss = cls_criterion(cls_logits, labels)
        total_cls_loss += cls_loss.item() * imgs.size(0)

        # 예측
        preds = torch.argmax(cls_logits, dim=1)
        all_preds.extend(preds.cpu().tolist())
        all_labels.extend(labels.cpu().tolist())

        # 틀린 이미지 저장
        for i in range(len(labels)):
            if preds[i] != labels[i]:
                wrong_images.append(paths[i])

# 평균 loss
avg_cls_loss = total_cls_loss / len(dataset)

# ==============================
# 6. 결과 출력
# ==============================
print(f"Classification Loss: {avg_cls_loss:.4f}")
accuracy = sum([p==l for p,l in zip(all_preds, all_labels)]) / len(all_labels)
print(f"Accuracy: {accuracy:.4f}")

print("\n틀린 이미지 예시 (처음 10개):")
for w in wrong_images[:10]:
    print(w)

# ==============================
# 7. 틀린 이미지 시각화 (옵션)
# ==============================
plt.figure(figsize=(12,6))
for i, img_path in enumerate(wrong_images[:8]):
    img = Image.open(img_path).convert("RGB")
    plt.subplot(2,4,i+1)
    plt.imshow(img)
    plt.title(f"Wrong")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python

import os
import json
import argparse
from pathlib import Path
from typing import List, Dict, Any, Tuple
import random
import math
import time

import numpy as np
from PIL import Image
import cv2
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from torchvision.models.segmentation import deeplabv3_resnet50
from torch.utils.tensorboard import SummaryWriter

from pycocotools.coco import COCO
from pycocotools import mask as maskUtils
from sklearn.metrics import roc_auc_score

def coco_polys_to_mask(h: int, w: int, ann_list: list) -> np.ndarray:
    rles = []
    for ann in ann_list:
        seg = ann.get("segmentation", [])
        if isinstance(seg, list) and len(seg) > 0 and isinstance(seg[0], str):
            seg = [list(map(float, s.split(","))) for s in seg]
        if len(seg) == 0:
            continue
        rle = maskUtils.frPyObjects(seg, h, w)
        if ann.get("iscrowd", 0) == 0:
            rle = maskUtils.merge(rle)
        rles.append(rle)
    if len(rles) == 0:
        return np.zeros((h, w), dtype=np.uint8)
    merged = maskUtils.merge(rles)
    m = maskUtils.decode(merged)
    if m.ndim == 3:
        m = np.any(m, axis=2)
    return (m > 0).astype(np.uint8)

def build_image_level_label(coco: COCO, img_id: int, cat_id_to_cls: Dict[int,int]) -> int:
    """Return 0=background, 1=benign, 2=malignant"""
    ann_ids = coco.getAnnIds(imgIds=[img_id])
    anns = coco.loadAnns(ann_ids)
    has_b, has_m = False, False
    for a in anns:
        c = cat_id_to_cls.get(a["category_id"], 0)
        if c == 1:
            has_b = True
        elif c == 2:
            has_m = True
    if has_m:
        return 2
    elif has_b:
        return 1
    else:
        return 0

# ----------------------
# Dataset
# ----------------------
class COCOMultiTaskDataset(Dataset):
    def __init__(self, root_dir: str, ann_file: str, img_size: int = 512,
                 augment: bool = False, cache_masks_dir: str = None):
        self.root_dir = Path(root_dir)
        self.coco = COCO(ann_file)
        self.img_ids = self.coco.getImgIds()
        self.img_size = img_size
        self.augment = augment
        self.cache_masks_dir = Path(cache_masks_dir) if cache_masks_dir else None
        if self.cache_masks_dir:
            self.cache_masks_dir.mkdir(parents=True, exist_ok=True)

        # map coco category id to our cls: benign=1, malignant=2
        cats = self.coco.loadCats(self.coco.getCatIds())
        self.cat_id_to_cls = {}
        for c in cats:
            name = c["name"].lower()
            if "benign" in name:
                self.cat_id_to_cls[c["id"]] = 1
            elif "malig" in name or "cancer" in name:
                self.cat_id_to_cls[c["id"]] = 2
            else:
                self.cat_id_to_cls[c["id"]] = 1

        self.img_trans = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
        ])

    def __len__(self):
        return len(self.img_ids)

    def _read_image(self, info: Dict[str,Any]) -> Image.Image:
        fname = info["file_name"]
        path = self.root_dir / fname
        if not path.exists():
            # images/ 폴더에 있을 수도 있음
            alt = self.root_dir / "images" / fname
            if alt.exists():
                path = alt
            else:
                raise FileNotFoundError(f"Image not found: {path} or {alt}")
        img = Image.open(path).convert("RGB")
        return img, str(path)

    def _get_mask_path(self, image_info):
        if not self.cache_masks_dir:
            return None
        base = Path(image_info["file_name"]).stem
        return self.cache_masks_dir / f"{base}.png"

    def _ensure_mask(self, img_info, anns, h, w):
        mask_path = self._get_mask_path(img_info)
        if mask_path and mask_path.exists():
            m = np.array(Image.open(mask_path).convert("L"))
            return (m>0).astype(np.uint8)
        lesion_anns = [a for a in anns if self.cat_id_to_cls.get(a["category_id"],0) in (1,2)]
        m = coco_polys_to_mask(h, w, lesion_anns)
        if mask_path:
            Image.fromarray((m*255).astype(np.uint8)).save(mask_path)
        return m

    def _resize_keep_ratio(self, img: Image.Image, mask: np.ndarray, size: int):
        w, h = img.size
        scale = size / max(h, w)
        nh, nw = int(round(h*scale)), int(round(w*scale))
        img_r = img.resize((nw, nh), Image.BILINEAR)
        mask_r = cv2.resize(mask.astype(np.uint8), (nw, nh), interpolation=cv2.INTER_NEAREST)
        pad_h = size - nh
        pad_w = size - nw
        img_p = Image.new("RGB", (size, size))
        img_p.paste(img_r, (pad_w//2, pad_h//2))
        mask_p = np.zeros((size, size), dtype=np.uint8)
        mask_p[pad_h//2:pad_h//2+nh, pad_w//2:pad_w//2+nw] = mask_r
        return img_p, mask_p

    def to_tensor(self, img: Image.Image):
        return self.img_trans(img)

    def __getitem__(self, idx):
        img_id = self.img_ids[idx]
        info = self.coco.loadImgs([img_id])[0]
        img, _ = self._read_image(info)
        h, w = info["height"], info["width"]

        ann_ids = self.coco.getAnnIds(imgIds=[img_id])
        anns = self.coco.loadAnns(ann_ids)

        # binary mask
        mask_full = self._ensure_mask(info, anns, h, w)

        # image-level label
        img_label = build_image_level_label(self.coco, img_id, self.cat_id_to_cls)

        # resize/pad
        img_sq, mask_sq = self._resize_keep_ratio(img, mask_full, self.img_size)

        img_t = self.to_tensor(img_sq)
        mask_t = torch.from_numpy(mask_sq).long()
        cls_t = torch.tensor(img_label-1, dtype=torch.long)  # convert 1/2 -> 0/1

        return img_t, mask_t, cls_t, info.get("file_name", str(img_id))


# ----------------------
# Model
# ----------------------
class MultiTaskDeepLab(nn.Module):
    def __init__(self, num_seg_classes:int=2, num_cls_classes:int=2, freeze_until: str = None):
        super().__init__()
        self.seg_model = deeplabv3_resnet50(weights="COCO_WITH_VOC_LABELS_V1", aux_loss=True)
        in_ch = self.seg_model.classifier[-1].in_channels
        self.seg_model.classifier[-1] = nn.Conv2d(in_ch, num_seg_classes, kernel_size=1)
        self.backbone = self.seg_model.backbone

        if freeze_until:
            for name,p in self.backbone.named_parameters():
                p.requires_grad = True
                if freeze_until == "layer3":
                    if not (name.startswith("layer3") or name.startswith("layer4")):
                        p.requires_grad = False

        self.cls_gap = nn.AdaptiveAvgPool2d((1,1))
        self.cls_head = nn.Sequential(
            nn.Flatten(),
            nn.Linear(2048,512),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(512,num_cls_classes)
        )

    def forward(self, x):
        seg_out = self.seg_model(x)["out"]
        feats = self.backbone(x)["out"]
        pooled = self.cls_gap(feats)
        cls_logits = self.cls_head(pooled)
        return seg_out, cls_logits
```

```python
import os
from pathlib import Path
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from torchvision.models.segmentation import deeplabv3_resnet50
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pycocotools.coco import COCO
from pycocotools import mask as maskUtils

# -----------------------------
# Helper functions
# -----------------------------
def coco_polys_to_mask(h: int, w: int, ann_list: list) -> np.ndarray:
    rles = []
    for ann in ann_list:
        seg = ann.get("segmentation", [])
        if isinstance(seg, list) and len(seg) > 0 and isinstance(seg[0], str):
            seg = [list(map(float, s.split(","))) for s in seg]
        if len(seg) == 0:
            continue
        rle = maskUtils.frPyObjects(seg, h, w)
        if ann.get("iscrowd", 0) == 0:
            rle = maskUtils.merge(rle)
        rles.append(rle)
    if len(rles) == 0:
        return np.zeros((h, w), dtype=np.uint8)
    merged = maskUtils.merge(rles)
    m = maskUtils.decode(merged)
    if m.ndim == 3:
        m = np.any(m, axis=2)
    return (m > 0).astype(np.uint8)

def build_image_level_label(coco: COCO, img_id: int, cat_id_to_cls: dict) -> int:
    ann_ids = coco.getAnnIds(imgIds=[img_id])
    anns = coco.loadAnns(ann_ids)
    has_b, has_m = False, False
    for a in anns:
        c = cat_id_to_cls.get(a["category_id"], 0)
        if c == 1:
            has_b = True
        elif c == 2:
            has_m = True
    if has_m:
        return 2
    elif has_b:
        return 1
    else:
        return 0

# -----------------------------
# Dataset
# -----------------------------
class COCOMultiTaskDataset(Dataset):
    def __init__(self, root_dir: str, ann_file: str, img_size: int = 512,
                 augment: bool = False, cache_masks_dir: str = None):
        self.root_dir = Path(root_dir)
        self.coco = COCO(ann_file)
        self.img_ids = self.coco.getImgIds()
        self.img_size = img_size
        self.augment = augment
        self.cache_masks_dir = Path(cache_masks_dir) if cache_masks_dir else None
        if self.cache_masks_dir:
            self.cache_masks_dir.mkdir(parents=True, exist_ok=True)

        cats = self.coco.loadCats(self.coco.getCatIds())
        self.cat_id_to_cls = {}
        for c in cats:
            name = c["name"].lower()
            if "benign" in name:
                self.cat_id_to_cls[c["id"]] = 1
            elif "malig" in name or "cancer" in name:
                self.cat_id_to_cls[c["id"]] = 2
            else:
                self.cat_id_to_cls[c["id"]] = 1

        self.img_trans = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
        ])

    def __len__(self):
        return len(self.img_ids)

    def _read_image(self, info: dict):
        fname = info["file_name"]
        path = self.root_dir / fname
        if not path.exists():
            alt = self.root_dir / "images" / fname
            if alt.exists():
                path = alt
            else:
                raise FileNotFoundError(f"Image not found: {path} or {alt}")
        img = Image.open(path).convert("RGB")
        return img, str(path)

    def _get_mask_path(self, image_info):
        if not self.cache_masks_dir:
            return None
        base = Path(image_info["file_name"]).stem
        return self.cache_masks_dir / f"{base}.png"

    def _ensure_mask(self, img_info, anns, h, w):
        mask_path = self._get_mask_path(img_info)
        if mask_path and mask_path.exists():
            m = np.array(Image.open(mask_path).convert("L"))
            return (m>0).astype(np.uint8)
        lesion_anns = [a for a in anns if self.cat_id_to_cls.get(a["category_id"],0) in (1,2)]
        m = coco_polys_to_mask(h, w, lesion_anns)
        if mask_path:
            Image.fromarray((m*255).astype(np.uint8)).save(mask_path)
        return m

    def _resize_keep_ratio(self, img: Image.Image, mask: np.ndarray, size: int):
        w, h = img.size
        scale = size / max(h, w)
        nh, nw = int(round(h*scale)), int(round(w*scale))
        img_r = img.resize((nw, nh), Image.BILINEAR)
        mask_r = cv2.resize(mask.astype(np.uint8), (nw, nh), interpolation=cv2.INTER_NEAREST)
        pad_h = size - nh
        pad_w = size - nw
        img_p = Image.new("RGB", (size, size))
        img_p.paste(img_r, (pad_w//2, pad_h//2))
        mask_p = np.zeros((size, size), dtype=np.uint8)
        mask_p[pad_h//2:pad_h//2+nh, pad_w//2:pad_w//2+nw] = mask_r
        return img_p, mask_p

    def to_tensor(self, img: Image.Image):
        return self.img_trans(img)

    def __getitem__(self, idx):
        img_id = self.img_ids[idx]
        info = self.coco.loadImgs([img_id])[0]
        img, path = self._read_image(info)
        h, w = info["height"], info["width"]

        ann_ids = self.coco.getAnnIds(imgIds=[img_id])
        anns = self.coco.loadAnns(ann_ids)

        mask_full = self._ensure_mask(info, anns, h, w)
        img_label = build_image_level_label(self.coco, img_id, self.cat_id_to_cls)
        img_sq, mask_sq = self._resize_keep_ratio(img, mask_full, self.img_size)
        img_t = self.to_tensor(img_sq)
        mask_t = torch.from_numpy(mask_sq).long()
        cls_t = torch.tensor(img_label-1, dtype=torch.long)

        return img_t, mask_t, cls_t, path

# -----------------------------
# 모델
# -----------------------------
class MultiTaskDeepLab(nn.Module):
    def __init__(self, num_seg_classes:int=2, num_cls_classes:int=2, freeze_until: str = None):
        super().__init__()
        self.seg_model = deeplabv3_resnet50(weights="COCO_WITH_VOC_LABELS_V1", aux_loss=True)
        in_ch = self.seg_model.classifier[-1].in_channels
        self.seg_model.classifier[-1] = nn.Conv2d(in_ch, num_seg_classes, kernel_size=1)
        self.backbone = self.seg_model.backbone

        if freeze_until:
            for name,p in self.backbone.named_parameters():
                p.requires_grad = True
                if freeze_until == "layer3":
                    if not (name.startswith("layer3") or name.startswith("layer4")):
                        p.requires_grad = False

        self.cls_gap = nn.AdaptiveAvgPool2d((1,1))
        self.cls_head = nn.Sequential(
            nn.Flatten(),
            nn.Linear(2048,512),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(512,num_cls_classes)
        )

    def forward(self, x):
        seg_out = self.seg_model(x)["out"]
        feats = self.backbone(x)["out"]
        pooled = self.cls_gap(feats)
        cls_logits = self.cls_head(pooled)
        return seg_out, cls_logits

# -----------------------------
# 시각화
# -----------------------------
def visualize_segmentation_comparison(orig_pil, true_mask, pred_mask, save_path):
    orig = np.array(orig_pil)
    H, W = orig.shape[:2]
    pred_mask_resized = cv2.resize(pred_mask.astype(np.uint8), (W,H), interpolation=cv2.INTER_NEAREST)
    true_mask_resized = cv2.resize(true_mask.astype(np.uint8), (W,H), interpolation=cv2.INTER_NEAREST)

    fig, axes = plt.subplots(1,4,figsize=(16,5))
    axes[0].imshow(orig); axes[0].set_title("Original"); axes[0].axis("off")
    axes[1].imshow(true_mask_resized, cmap="gray"); axes[1].set_title("Ground Truth"); axes[1].axis("off")
    axes[2].imshow(pred_mask_resized, cmap="gray"); axes[2].set_title("Prediction"); axes[2].axis("off")
    
    overlay = orig.copy()
    overlay[pred_mask_resized==1] = [255,0,0]
    blended = (0.7*orig + 0.3*overlay).astype(np.uint8)
    axes[3].imshow(blended); axes[3].set_title("Overlay"); axes[3].axis("off")
    
    plt.tight_layout()
    plt.savefig(save_path)
    plt.close()

# -----------------------------
# 평가 + 틀린 샘플 저장 + Accuracy / Loss
# -----------------------------
def evaluate_model(model, loader, device, save_dir="misclassified_compare", max_samples=20):
    os.makedirs(save_dir, exist_ok=True)
    wrong_count = 0
    class_names = ["benign", "malignant"]
    criterion_seg = nn.CrossEntropyLoss()
    criterion_cls = nn.CrossEntropyLoss()

    total_loss = 0.0
    total_correct = 0
    total_samples = 0

    model.eval()
    with torch.no_grad():
        for imgs, masks, cls_labels, paths in loader:
            imgs, masks, cls_labels = imgs.to(device), masks.to(device), cls_labels.to(device)
            seg_logits, cls_logits = model(imgs)
            preds_cls = torch.argmax(F.softmax(cls_logits, dim=1), dim=1)
            preds_seg = torch.argmax(seg_logits, dim=1)

            # Loss 계산
            loss_seg = criterion_seg(seg_logits, masks)
            loss_cls = criterion_cls(cls_logits, cls_labels)
            batch_loss = loss_seg + loss_cls
            total_loss += batch_loss.item() * imgs.size(0)

            # Accuracy 계산
            total_correct += (preds_cls == cls_labels).sum().item()
            total_samples += imgs.size(0)

            # 틀린 샘플 저장
            for i in range(len(preds_cls)):
                if preds_cls[i].item() != cls_labels[i].item():
                    wrong_count += 1
                    if wrong_count > max_samples:
                        continue
                    orig = Image.open(paths[i]).convert("RGB")
                    save_path = Path(save_dir)/f"wrong_{wrong_count}_{Path(paths[i]).name}_true{class_names[cls_labels[i].item()]}_pred{class_names[preds_cls[i].item()]}.jpg"
                    visualize_segmentation_comparison(orig, masks[i].cpu().numpy(), preds_seg[i].cpu().numpy(), save_path)
                    print(f"[{wrong_count}] Saved: {save_path} | True: {class_names[cls_labels[i].item()]} | Pred: {class_names[preds_cls[i].item()]}")

    avg_loss = total_loss / total_samples
    acc = total_correct / total_samples
    print(f"총 {wrong_count}개의 잘못된 예측 이미지 저장 완료")
    print(f"Classification Accuracy: {acc*100:.2f}%, Average Loss: {avg_loss:.4f}")

# -----------------------------
# 실행
# -----------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = MultiTaskDeepLab(num_seg_classes=2, num_cls_classes=2).to(device)
ckpt = torch.load("../runs_multitask_1000_ver2/best_multitask.pt", map_location=device)
model.load_state_dict(ckpt["model"], strict=False)

test_dir = "../dataset_1000/test"
test_ann = f"{test_dir}/_annotations.coco.json"
ds_test = COCOMultiTaskDataset(root_dir=test_dir, ann_file=test_ann, img_size=512, augment=False, cache_masks_dir="masks_cache")

def collate(batch):
    imgs, masks, clss, paths = zip(*batch)
    return torch.stack(imgs), torch.stack(masks), torch.stack(clss), list(paths)

dl_test = DataLoader(ds_test, batch_size=4, shuffle=False, collate_fn=collate)

evaluate_model(model, dl_test, device, save_dir="../misclassified_compare", max_samples=30)
```

```python
import os
from pathlib import Path
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from torchvision.models.segmentation import deeplabv3_resnet50

# -----------------------------
# 1. CSV 기반 Dataset
# -----------------------------
class CsvImageDataset(Dataset):
    def __init__(self, csv_path, img_root, img_size=512):
        self.df = pd.read_csv(csv_path)
        self.img_root = Path(img_root)
        self.img_size = img_size
        self.transform = transforms.Compose([
            transforms.Resize((img_size,img_size)),
            transforms.ToTensor()
        ])
        # 문자열 라벨 → 숫자 인덱스
        self.class_names = ["Benign", "Malignant", "Indeterminate"]
        self.label2idx = {name: idx for idx, name in enumerate(self.class_names)}
        # CSV 컬럼명
        self.img_col = 'isic_id'  # ISIC 2016 CSV 기준
        self.label_col = 'diagnosis_1'   # 0=benign,1=malignant 등 실제 컬럼 확인 필요

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_path = self.img_root / f"{row[self.img_col]}.jpg"
        label_str = row[self.label_col]
        # 없는 라벨 대비 안전하게
        label = self.label2idx.get(label_str, -1)  
        if label == -1:
            raise ValueError(f"Unknown label '{label_str}' at row {idx}")
        img = Image.open(img_path).convert("RGB")
        img_t = self.transform(img)
        return img_t, label, str(img_path)



# -----------------------------
# 2. MultiTask 모델 정의
# -----------------------------
class MultiTaskDeepLab(nn.Module):
    def __init__(self, num_seg_classes:int=2, num_cls_classes:int=2):
        super().__init__()
        self.seg_model = deeplabv3_resnet50(weights="COCO_WITH_VOC_LABELS_V1", aux_loss=True)
        in_ch = self.seg_model.classifier[-1].in_channels
        self.seg_model.classifier[-1] = nn.Conv2d(in_ch, num_seg_classes, kernel_size=1)
        self.backbone = self.seg_model.backbone
        self.cls_gap = nn.AdaptiveAvgPool2d((1,1))
        self.cls_head = nn.Sequential(
            nn.Flatten(),
            nn.Linear(2048,512),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(512,num_cls_classes)
        )

    def forward(self, x):
        seg_out = self.seg_model(x)["out"]
        feats = self.backbone(x)["out"]
        pooled = self.cls_gap(feats)
        cls_logits = self.cls_head(pooled)
        return seg_out, cls_logits

# -----------------------------
# 3. 시각화
# -----------------------------
def visualize_pred(orig_pil, pred_cls, pred_mask, true_cls, save_path):
    orig = np.array(orig_pil)
    H,W = orig.shape[:2]
    pred_mask_resized = cv2.resize(pred_mask.astype(np.uint8), (W,H), interpolation=cv2.INTER_NEAREST)

    fig, axes = plt.subplots(1,3,figsize=(14,5))
    axes[0].imshow(orig); axes[0].set_title("Original"); axes[0].axis("off")
    axes[1].imshow(pred_mask_resized, cmap="gray"); axes[1].set_title("Pred Mask"); axes[1].axis("off")

    overlay = orig.copy()
    overlay[pred_mask_resized==1] = [255,0,0]
    blended = (0.7*orig + 0.3*overlay).astype(np.uint8)
    axes[2].imshow(blended); axes[2].set_title(f"Pred: {pred_cls}, True: {true_cls}"); axes[2].axis("off")

    plt.tight_layout()
    plt.savefig(save_path)
    plt.close()

# -----------------------------
# 4. 전체 예측 + 틀린 샘플 시각화 + Acc/Loss
# -----------------------------
def evaluate_csv_model(model, dataset, device, save_dir="misclassified_results", max_wrong=30, batch_size=4):
    os.makedirs(save_dir, exist_ok=True)
    loader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

    criterion_cls = nn.CrossEntropyLoss()
    model.eval()

    total_loss = 0
    total_correct = 0
    total_samples = 0
    wrong_count = 0

    with torch.no_grad():
        for imgs, labels, paths in loader:
            imgs = imgs.to(device)
            labels = torch.tensor(labels).to(device)

            seg_logits, cls_logits = model(imgs)
            preds_cls = torch.argmax(cls_logits, dim=1)

            loss_cls = criterion_cls(cls_logits, labels)
            batch_size_actual = imgs.size(0)
            total_loss += loss_cls.item() * batch_size_actual
            total_correct += (preds_cls == labels).sum().item()
            total_samples += batch_size_actual

            # 틀린 샘플 시각화
            for i in range(batch_size_actual):
                if preds_cls[i].item() != labels[i].item():
                    wrong_count += 1
                    if wrong_count > max_wrong:
                        continue
                    orig = Image.open(paths[i]).convert("RGB")
                    # Segmentation mask는 예측만
                    pred_mask = torch.argmax(seg_logits[i], dim=0).cpu().numpy()
                    save_path = Path(save_dir) / f"wrong_{wrong_count}_true{labels[i].item()}_pred{preds_cls[i].item()}.jpg"
                    visualize_segmentation_comparison(orig, pred_mask, save_path)
                    print(f"[{wrong_count}] Saved: {save_path}")

    acc = total_correct / total_samples
    avg_loss = total_loss / total_samples
    print(f"Classification Accuracy: {acc:.4f}, Avg Loss: {avg_loss:.4f}")

# -----------------------------
# 5. 실행 예시
# -----------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 모델 로드
model = MultiTaskDeepLab(num_seg_classes=2, num_cls_classes=2)
ckpt = torch.load("../runs_multitask_1000_ver2/best_multitask.pt", map_location=device)
model.load_state_dict(ckpt["model"], strict=False)
model.to(device)

# CSV 데이터셋
dataset = CsvImageDataset(csv_path="../ISIC_dataset/challenge-2016-test_metadata_2025-09-01.csv", img_root="../ISIC_dataset/img")

# 평가
evaluate_csv_model(model, dataset, device, save_dir="misclassified_results_ISIC", max_wrong=30)
```

```python
import os
from pathlib import Path
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from torchvision.models.segmentation import deeplabv3_resnet50
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# ----------------------------
# CSV Dataset
# ----------------------------
class CsvImageDataset(Dataset):
    def __init__(self, csv_path, img_root, transform=None, label_map=None):
        self.df = pd.read_csv(csv_path)
        self.img_root = Path(img_root)
        self.transform = transform if transform else transforms.Compose([
            transforms.Resize((512,512)),
            transforms.ToTensor(),
            transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
        ])
        self.label_col = "diagnosis_1"
        self.label2idx = {"Benign":0, "Malignant":1, "Indeterminate":1} if label_map is None else label_map
        self.df['img_path'] = self.df['isic_id'].apply(self._find_file)

    def _find_file(self, name):
        for ext in ['.jpg', '.png', '.jpeg']:
            path = self.img_root / f"{name}{ext}"
            if path.exists():
                return str(path)
        raise FileNotFoundError(f"Image file not found for {name} in {self.img_root}")

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_path = row['img_path']
        label_str = row[self.label_col]
        label = self.label2idx.get(label_str, 1)
        img = Image.open(img_path).convert("RGB")
        img_t = self.transform(img)
        return img_t, label, img_path

# ----------------------------
# Multi-task DeepLab Model
# ----------------------------
class MultiTaskDeepLab(nn.Module):
    def __init__(self, num_seg_classes:int=2, num_cls_classes:int=2):
        super().__init__()
        self.seg_model = deeplabv3_resnet50(weights="COCO_WITH_VOC_LABELS_V1", aux_loss=True)
        in_ch_seg = self.seg_model.classifier[-1].in_channels
        self.seg_model.classifier[-1] = nn.Conv2d(in_ch_seg, num_seg_classes, kernel_size=1)

        # 분류용 head: seg_out의 채널 수를 Linear 입력으로 사용
        self.cls_head = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Linear(num_seg_classes, num_cls_classes)  # ← 수정: seg_out 채널 수
        )

    def forward(self, x):
        seg_out = self.seg_model(x)['out']  # (B, num_seg_classes, H, W)
        cls_out = self.cls_head(seg_out)    # AdaptiveAvgPool2d + Flatten
        return seg_out, cls_out

# ----------------------------
# Segmentation Overlay 시각화
# ----------------------------
def visualize_segmentation_comparison(orig_pil, pred_mask, save_path, true_label=None, pred_label=None):
    orig = np.array(orig_pil)
    overlay = orig.copy()
    if pred_mask.shape != orig.shape[:2]:
        pred_mask = np.array(Image.fromarray(pred_mask.astype(np.uint8)).resize((orig.shape[1], orig.shape[0]), Image.NEAREST))
    overlay[pred_mask==1] = [255,0,0]
    blended = (0.7*orig + 0.3*overlay).astype(np.uint8)

    fig, axes = plt.subplots(1,2,figsize=(10,5))
    axes[0].imshow(orig); axes[0].set_title("Original"); axes[0].axis("off")
    axes[1].imshow(blended); axes[1].set_title(f"Prediction Overlay\nTrue:{true_label}, Pred:{pred_label}" if true_label is not None else "Prediction Overlay"); axes[1].axis("off")
    plt.tight_layout()
    plt.savefig(save_path)
    plt.close()

# ----------------------------
# 평가 함수
# ----------------------------
def evaluate_csv_model(model, dataset, device, save_dir="results", max_wrong=30, max_correct=40):
    os.makedirs(save_dir, exist_ok=True)
    loader = DataLoader(dataset, batch_size=4, shuffle=False, num_workers=4, pin_memory=True)

    criterion_cls = nn.CrossEntropyLoss()

    model.eval()
    total_correct = 0
    total_samples = 0
    total_loss = 0
    wrong_count = 0
    correct_count = 0
    misclass_counts = {"Benign_to_Malignant":0, "Malignant_to_Benign":0}

    with torch.no_grad():
        for imgs, labels, paths in loader:
            imgs = imgs.to(device)
            labels_tensor = torch.as_tensor(labels, dtype=torch.long, device=device)

            seg_logits, cls_logits = model(imgs)

            # Classification loss
            loss_cls = criterion_cls(cls_logits, labels_tensor)
            total_loss += loss_cls.item() * imgs.size(0)

            preds_cls = torch.argmax(F.softmax(cls_logits, dim=1), dim=1)
            total_correct += (preds_cls == labels_tensor).sum().item()
            total_samples += imgs.size(0)

            # 샘플별 처리
            for i in range(len(labels)):
                true_label = labels[i]
                pred_label = preds_cls[i].item()
                pred_mask = torch.argmax(seg_logits[i], dim=0).cpu().numpy()
                orig = Image.open(paths[i]).convert("RGB")

                # 오분류 카운트
                if true_label == 0 and pred_label == 1:
                    misclass_counts["Benign_to_Malignant"] += 1
                elif true_label == 1 and pred_label == 0:
                    misclass_counts["Malignant_to_Benign"] += 1

                # 틀린 샘플 저장
                if pred_label != true_label and wrong_count < max_wrong:
                    wrong_count += 1
                    save_path = Path(save_dir) / f"wrong_{wrong_count}_true{true_label}_pred{pred_label}.jpg"
                    visualize_segmentation_comparison(orig, pred_mask, save_path, true_label, pred_label)
                # 맞춘 샘플 저장
                elif pred_label == true_label and correct_count < max_correct:
                    correct_count += 1
                    save_path = Path(save_dir) / f"correct_{correct_count}_label{true_label}.jpg"
                    visualize_segmentation_comparison(orig, pred_mask, save_path, true_label, pred_label)

    acc = total_correct / total_samples
    avg_loss = total_loss / total_samples
    benign_to_mal_ratio = misclass_counts["Benign_to_Malignant"] / total_samples
    mal_to_ben_ratio = misclass_counts["Malignant_to_Benign"] / total_samples

    print(f"Accuracy: {acc:.4f}, Avg Loss: {avg_loss:.4f}")
    print(f"Benign → Malignant: {benign_to_mal_ratio:.3f}, Malignant → Benign: {mal_to_ben_ratio:.3f}")
    print(f"Saved {wrong_count} wrong samples, {correct_count} correct samples.")
    return acc, avg_loss

# ----------------------------
# 사용 예제
# ----------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = MultiTaskDeepLab(num_seg_classes=2, num_cls_classes=2).to(device)
ckpt = torch.load("../runs_multitask_1000/best_multitask.pt", map_location=device)
model.load_state_dict(ckpt["model"], strict=False)
model.eval()

dataset = CsvImageDataset(
    csv_path="../ISIC_dataset/challenge-2016-test_metadata_2025-09-01.csv",
    img_root="../ISIC_dataset/img"
)

acc, avg_loss = evaluate_csv_model(
    model, dataset, device,
    save_dir="results_with_correct_fixed",
    max_wrong=len(dataset),
    max_correct=40
)
```
